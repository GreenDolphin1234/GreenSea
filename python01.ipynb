# filter와 map

```python
# filter는 true인 값만 남김
data = [1, -2, 3, 4, -5, 0, -1, 3,-4, 5]

def pos(v):
    return v > 0  # 양수인 경우에만 True를 반환

print(list(filter(pos, data)))  # filter를 사용하여 양수 필터링
```

```python
data = [1, -2, 3, 4, -5, 0, -1, 3,-4, 5]
pos = [True, False, True, True, False, False, False, True, False, True]
print(list(filter(lambda v: v > 0, data)))  # 람다 함수를 사용하여 양수 필터링  
```

```python
data = [1, -2, 3, 4, -5, 0, -1, 3,-4, 5]
def pow(v):
    return v ** 2  # 제곱을 반환
print(list(map(pow, data)))  # map을 사용하여 제곱 계산
```

```python
data = [1, -2, 3, 4, -5, 0, -1, 3,-4, 5]
print(list(map(lambda v: v ** 2, data)))  # 람다 함수를 사용하여 제곱 계산
```

```python
lis = list(range(1, 10))  #1부터 9까지의 숫자를 포함하는 리스트 생성   

def odd(x):
    return x % 2 == 1  # 홀수인 경우에만 True를 반환    

def even(x):
    return x % 2 == 0  # 짝수인 경우에만 True를 반환

def add(x):
    return x + 10 # 10을 더한 값을 반환

def mul(x):
    return x * 10 # 10을 곱한 값을 반환 

def addmul(x):
    return x + 10, x * 10 # 10을 더하고 곱한 값을 반환

print(list(filter(odd, lis)))  # 홀수 필터링
print(list(filter(even, lis)))  # 짝수 필터링   
print(list(map(add, lis)))  # 10을 더한 값
print(list(map(mul, lis)))  # 10을 곱한 값  
print(list(map(addmul, lis)))  # 10을 더하고 곱한 값
```

```python
import itertools

students = ['학생1', '학생2', '학생3', '학생4', '학생5']
snacks = ['과자', '음료수', '초콜릿', '사탕']

result = zip(students, snacks)  # 학생과 간식을 쌍으로 묶기
print(list(result))  # 결과를 리스트로 변환하여 출력

result2 = itertools.zip_longest(students, snacks, fillvalue='칸초')  # 학생과 간식을 쌍으로 묶기 (길이가 다른 경우 '없음'으로 채움)
print(list(result2))  # 결과를 리스트로 변환하여 출력
```
